```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification

# Choose a pretrained model and tokenizer for sentiment analysis
model_name = "nlptown/bert-base-multilingual-uncased-sentiment"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSequenceClassification.from_pretrained(model_name)

# Sample text for sentiment analysis
text = "I love using Hugging Face's pretrained models for natural language processing!"

# Tokenize the text
tokens = tokenizer(text, return_tensors="pt")

# Forward pass through the model
outputs = model(**tokens)

# Get the predicted sentiment label
predicted_label = outputs.logits.argmax().item()

# Decode the tokenized input back to text
decoded_text = tokenizer.decode(tokens["input_ids"][0])

# Print results
print("Original Text:", text)
print("Tokenized Input:", tokens)
print("Decoded Text:", decoded_text)
print("Predicted Sentiment Label:", predicted_label)
```
